using simple imputer to automate the process of filling null values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
numeric_df = df.select_dtypes(include=['number'])
imputer.fit(numeric_df)
X = imputer.transform(df)
```

```python
df = pd.DataFrame(X,columns=df.columns)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(df,test_size=0.2,random_state=42)
```

here we notice that chas has very less 1 counts and we can go in a problem if there is no 1 in our training set so we will use StratifiedShuffleSplit instead so that we could encounter this problem

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(df,df['CHAS']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_train_set['CHAS'].value_counts()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',StandardScaler()),
])
```

```python
df = my_pipeline.fit_transform(df)
```

```python
X = strat_train_set.drop('MEDV',axis=1)
y = strat_train_set['MEDV']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(X,y)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X)
```

```python
import numpy as np
np.sqrt(mean_squared_error(y,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=10)
rmse_scores = np.sqrt(-score)
rmse_scores
```

```python
def print_scores(scores):
    print("Scores :",scores)
    print("Mean : ",scores.mean())
    print("Standard deviation : ",scores.std())
```

```python
print_scores(rmse_scores)
```

```python
print_scores(rmse_scores)
```
